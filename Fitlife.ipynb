```python
import pandas as pd

df = pd.read_csv(r"\\ad.wayne.edu\Users\ex4360\Documents\Health fitness\health_fitness_dataset.csv")
df.head()
```

```python
#Do people who take more steps burn more calories?

print(df[["daily_steps", "calories_burned"]].corr())

import matplotlib.pyplot as plt
plt.scatter(df["daily_steps"], df["calories_burned"], alpha=0.2)
plt.xlabel("Daily Steps")
plt.ylabel("Calories Burned")
plt.title("Steps vs Calories Burned")
plt.show()
```

```python
# Is there a link between sleep hours and activity levels?

print(df[["hours_sleep", "daily_steps"]].corr())

plt.scatter(df["hours_sleep"], df["daily_steps"], alpha=0.2)
plt.xlabel("Hours of Sleep")
plt.ylabel("Daily Steps")
plt.title("Sleep vs Activity")
plt.show()
```

```python
#Which days have the highest average activity
df["date"] = pd.to_datetime(df["date"])
df["weekday"] = df["date"].dt.day_name()

avg_steps_by_day = df.groupby("weekday")["daily_steps"].mean().sort_values(ascending=False)
print(avg_steps_by_day)

avg_steps_by_day.plot(kind="bar", title="Average Steps by Day of Week")
plt.ylabel("Steps")
plt.show()
```
